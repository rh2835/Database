This is a practice for setting up PostgreSQL and integrating PostgreSQL in Python

Some basic definition:<br/>
1. table vs schema: a relation schema is a logical definition of a table, it defines the name, attributes and type of each of columns. A database schema is the collection of relation schemas for a whole database. A table is a structure with bunch of rows, each of row has attributes defined by the schema.<br/>


2. ORDBMS(object-relational) vs RDBMS:<br/>
  * OR: objects are stored in tables of objects rather than in tables of rows
  * OR: provide a bridge between the relational and object-oriented paradigms

1. Installing Postgres by brew<br/>
brew install postgresql<br/>
brew services start postgresql<br/>
check version: postgres -V


2. Configuring Postgres<br/>
psql postgres<br/>
list all the users: postgres=# \du<br/>
create role: CREATE ROLE username WITH LOGIN PASSWORD 'quoted password' [OPTIONS]<br/>
set up password: postgres=# \password username<br/>
add permission: postgres=# ALTER ROLE username CREATEDB;<br/>


3. Creating a database<br/>
CREATE DATABASE databasename;<br/>
GRANT ALL PRIVILEGES ON DATABASE super_awesome_application TO username;<br/> 
list all databases: \list;<br/>
connect to specific database: \connect;<br/>
list tables in the currently connected database: \dt;<br/>
